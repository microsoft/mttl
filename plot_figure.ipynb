```python
import numpy as np 
adapter_embedding_dict = np.load("adapter_embedding_None.npy", allow_pickle=True).item()
adapter_embedding_dict
```

```python
qa = []
for key in adapter_embedding_dict.keys():
    if "qa" in key:
        qa.append(key)
qa  
```

```python
len(qa)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
embedding = []

for key in qa:
    embedding.append(adapter_embedding_dict[key])
embedding = np.array(embedding)    
embedding.shape

```

```python
embedding = np.squeeze(embedding, axis=1)
cosine_similarity(embedding)
```

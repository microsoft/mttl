```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd 

x = [540,2.2*1024,2.2*1024,3.5,8.6,2.2*1024,2.2*1024,1.4*1024,1.4*1024]

# size = [540, 84*1024, 2.2*1024, 50*1024, 17*1024, 17*1024,17*1024]
# marker_size = [0.54 , 2.2, 2.2, 35, 0.35§, 2.2, 0.35, 2.2]
y = [66.2,66.1,66.8,65.4,66.3,68.8,69.2, 69.3,68.0]

color_map = ["#1f77b4","#ff7f0e","#2ca02c","yellow","#d62728","#8c564b","#8c564b","yellow"]

name_text=["IA3","AdapterSoup","LoRA","Poly-z","TensorPoly-I-z","Poly","MHR", "TensorPoly-I", "TensorPoly-II"]

df = pd.DataFrame({"x":x,"ACC":y,"text":name_text})
df["color"] = df["ACC"].astype(str)

color_discrete_map= dict(zip(df["color"].to_list(), color_map))
print(color_discrete_map)
marker_symbol_list = ["circle", "square", "diamond-tall", "diamond", "hexagram", "cross-dot", "cross", "hexagram", "hexagram"]
```

```python


fig = px.scatter(df,x="x",y="ACC", color = "ACC", log_x = True,text="text",color_discrete_map=color_discrete_map)

# fig = px.scatter(x=x, y=y, |
#                  text=name_text, log_x=True, labels = name_text, 
#                  size = size,symbol=marker_symbol_list,
#                  color=name_text, size_max=60,
#                  )

fig.update_traces(textposition='middle right',
                  marker_line_width=2,marker_line_color="black", marker_symbol=marker_symbol_list)
fig.update_traces(
    # mode='markers',
    marker=dict(
        size=25,
        color=color_map,  # Array of colors
        symbol=marker_symbol_list,
    )
)

fig.update_layout(
    height=800,
    # title_text='GDP and Life Expectancy (Americas, 2007)'
)
fig.update_layout(showlegend=False)



fig.update_layout(
    xaxis = dict(
        tickmode = 'array',
        tickvals = [5, 500, 5000, 50000],
        ticktext = ['5K', '500K', '5M', "50M"],
        # showgrid=True,
    ),
    # yaxis=dict(showgrid=False)
    # fig.update_layout(xaxis=dict(showgrid=False),
    #           yaxis=dict(showgrid=False)
)

font_dict=dict(family='Arial',
               size=21,
               color='black'
               )


# general figure formatting
fig.update_layout(
                font=font_dict ,
                  plot_bgcolor='white',  # background color
                  width=850,  # figure width
                  height=700,  # figure height
                #   margin=dict(r=20,t=20,b=10)  # remove white space ,
     
            )

fig.update_yaxes(
                title_text='Accuracy',  # axis label
                 showline=True,  # add line at x=0
                 linecolor='black',  # line color
                 linewidth=2.4, # line size
                 ticks='outside',  # ticks outside axis
                 tickfont=font_dict, # tick label font
                 mirror='allticks',  # add ticks to top/right axes
                 tickwidth=2.4,  # tick width
                 tickcolor='black',  # tick color
                 )
fig.update_xaxes(title_text='Adaptation Parameters',
                 showline=True,
                #  showticklabels=True,
                 linecolor='black',
                 linewidth=2.4,
                 ticks='outside',
                 tickfont=font_dict,
                 mirror='allticks',
                 tickwidth=2.4,
                 tickcolor='black',
                 range = [0,5.5]
                 )

# fig.add_shape(type="circle",
#     xref="x", yref="y",
#     x0=1, y0=66,
#     x1=100, y1=max(y),
#     opacity=0.2,
#     fillcolor="orange",
#     line_color="orange",
# )

# fig.add_trace(go.Scatter(
#     x=[10],
#     y=[69.5],
#     text=["Routing-Tuning"],
#     mode="text",
#     textfont=dict(
#         color="black",
#         size=28,
#         family="Arail",
#     )
# ))
# fig.update_layout(plot_bgcolor="white")
fig.add_hline(y=69.2,line_width=3, line_dash="dash", line_color="black",
              annotation_text="SOTA results",
              annotation_position="bottom",
              annotation_font_size=20
              )
fig.show()
fig.write_image("figure/test_acc.pdf")
```
